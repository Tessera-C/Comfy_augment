```python
import os

def add_prefix_to_filenames(folder_path, prefix):
    if not os.path.isdir(folder_path):
        print(f"경로가 폴더가 아닙니다: {folder_path}")
        return

    for filename in os.listdir(folder_path):
        old_path = os.path.join(folder_path, filename)
        if os.path.isfile(old_path):
            new_filename = prefix + filename
            new_path = os.path.join(folder_path, new_filename)

            # 파일 이름이 이미 존재하지 않을 때만 변경
            if not os.path.exists(new_path):
                os.rename(old_path, new_path)
                print(f"'{filename}' → '{new_filename}' 이름 변경 완료")
            else:
                print(f"[경고] 파일 이름 충돌: {new_filename} 은 이미 존재합니다.")

# 예시 사용
# 폴더 경로에 있는 모든 파일 이름 앞에 'v1_' 을 붙임
folder = "output/TiROD_v15"
prefix = "v15_"
add_prefix_to_filenames(folder, prefix)
```

```python
import random
import os
import shutil

# 설정: 버전과 비율
version = "v9"
delete_ratio = 0.5  # 삭제 비율 (예: 0.5 -> 50% 삭제)
random_seed = 42    # 재현 가능한 시드

# 경로 설정
base_dir = "/home/jhcha/jh_ws/yolo/datasets"
source_folder = os.path.join(base_dir, "ODSR-IHS")
destination_folder = os.path.join(base_dir, f"ODSR-IHS_{version}_{int(delete_ratio * 100)}")

# 원본 폴더 복사
if os.path.exists(destination_folder):
    print(f"{destination_folder} 이미 존재함. 삭제 후 새로 복사합니다.")
    shutil.rmtree(destination_folder)
shutil.copytree(source_folder, destination_folder)
print(f"복사 완료: {destination_folder}")

# 랜덤 삭제 처리
train_dir = os.path.join(destination_folder, "train")

# 이미지 목록 (파일만 필터링)
image_files = [
    f for f in os.listdir(train_dir)
    if os.path.isfile(os.path.join(train_dir, f)) and f.lower().endswith(('.jpg', '.jpeg', '.png'))
]
print(f"총 이미지 파일 수: {len(image_files)}")

# 삭제할 파일 개수 계산
delete_count = int(len(image_files) * delete_ratio)
random.seed(random_seed)
files_to_delete = random.sample(image_files, delete_count)

# 삭제 실행
for img_file in files_to_delete:
    img_path = os.path.join(train_dir, img_file)
    label_path = os.path.join(train_dir, os.path.splitext(img_file)[0] + ".txt")

    try:
        if os.path.isfile(img_path):
            os.remove(img_path)
        if os.path.isfile(label_path):
            os.remove(label_path)
    except Exception as e:
        print(f"삭제 중 오류 발생: {e} (파일: {img_file})")

print(f"삭제 완료: {delete_count}개 파일 삭제 (비율 {delete_ratio*100}%)")
```

```python
import os
import shutil

# 설정
version = "v9"
append_str = f"{version}_"
image_folder = f"output/ODSR_{version}"
label_folder = f"output/ODSR_{version}_anno"

# 대상 train 폴더 경로
base_dir = "/home/jhcha/jh_ws/yolo/datasets"
train_dir = os.path.join(base_dir, f"ODSR-IHS_{version}_{int(delete_ratio * 100)}", "train")

# 기준: train 폴더 내에 존재하는 '기본 이름 파일' (00001.jpg 등)
existing_images = [
    f for f in os.listdir(train_dir)
    if os.path.isfile(os.path.join(train_dir, f)) and f.lower().endswith(('.jpg', '.jpeg', '.png')) and not f.startswith(append_str)
]
print(f"기준 파일 수 (접두어 없는 파일): {len(existing_images)}")

# 확장자 없이 basename 추출 (00001 등)
basenames = set(os.path.splitext(f)[0] for f in existing_images)

copy_count = 0
for basename in sorted(basenames):
    # 이미지 파일 찾기
    src_img_path = None
    for ext in ['.jpg', '.jpeg', '.png']:
        candidate = os.path.join(image_folder, append_str + basename + ext)
        if os.path.isfile(candidate):
            src_img_path = candidate
            dst_img_path = os.path.join(train_dir, append_str + basename + ext)
            shutil.copy2(src_img_path, dst_img_path)
            break

    # 라벨 파일 찾기 (.txt)
    src_label_path = os.path.join(label_folder, append_str + basename + ".txt")
    dst_label_path = os.path.join(train_dir, append_str + basename + ".txt")
    if os.path.isfile(src_label_path):
        shutil.copy2(src_label_path, dst_label_path)

    if src_img_path and os.path.isfile(src_label_path):
        copy_count += 1

print(f"복사 완료: {copy_count} 쌍의 v9_ 접두어 파일")
```

```python
import os
import yaml

# 설정
version = "v9"
delete_ratio = 0.5
version_tag = f"{version}_{int(delete_ratio * 100)}"
yaml_filename = f"{version_tag}.yaml"
yaml_path = os.path.join("/home/jhcha/jh_ws/yolo", yaml_filename)

dataset_folder = f"ODSR-IHS_{version_tag}"

# YAML 데이터 정의
yaml_content = {
    'train': f"{dataset_folder}/train",
    'val': f"{dataset_folder}/valid",
    'test': f"{dataset_folder}/test",
    'nc': 14,
    'names': ['slippers', 'stool', 'wire', 'carpet', 'sofa', 'socks', 'feces',
              'table', 'bed', 'closetool', 'book', 'cabinet', 'trashcan', 'curtain'],
    'roboflow': {
        'workspace': 'agv',
        'project': 'odsr-ihs',
        'version': 1,
        'license': 'CC BY 4.0',
        'url': 'https://universe.roboflow.com/agv/odsr-ihs/dataset/1'
    }
}

# YAML 저장 (인라인 리스트 형식)
with open(yaml_path, 'w') as f:
    yaml.dump(yaml_content, f, sort_keys=False, default_flow_style=False)

# 후처리: 'names'만 대괄호 형식으로 강제 치환
with open(yaml_path, 'r') as f:
    lines = f.readlines()

with open(yaml_path, 'w') as f:
    for line in lines:
        if line.startswith("names:"):
            f.write("names: ['slippers', 'stool', 'wire', 'carpet', 'sofa', 'socks', 'feces', 'table', 'bed', 'closetool', 'book', 'cabinet', 'trashcan', 'curtain']\n")
        elif line.strip().startswith("- "):
            continue  # 블록 스타일 리스트 제거
        else:
            f.write(line)

print(f"YAML 파일 생성 완료: {yaml_path}")
```

```python
import os
import subprocess

# 설정
version = "v9"
delete_ratio = 0.5
version_tag = f"{version}_{int(delete_ratio * 100)}"
yaml_name = f"{version_tag}.yaml"

# 경로
train_script_path = "/home/jhcha/jh_ws/yolo/yolo_train_ODSR_half.py"
working_dir = os.path.dirname(train_script_path)

# 1. 학습 스크립트 내부의 data="..." 줄 수정
with open(train_script_path, "r") as f:
    lines = f.readlines()

with open(train_script_path, "w") as f:
    for line in lines:
        if line.strip().startswith("data="):
            f.write(f'    data="{yaml_name}",\n')
        else:
            f.write(line)

print(f'✅ "data=" 경로를 {yaml_name}로 수정 완료\n')

# 2. 스크립트 실행 (작업 디렉토리 지정)
result = subprocess.run(
    ["python3", train_script_path],
    # capture_output=True,
    text=True,
    cwd=working_dir
)

# 3. 결과 출력
print("===== STDOUT =====")
print(result.stdout)
print("\n===== STDERR =====")
print(result.stderr)

if result.returncode == 0:
    print("\n✅ 훈련 스크립트 실행 완료")
else:
    print(f"\n❌ 실행 중 오류 발생 (return code: {result.returncode})")
```
